Load the data

```python
def parseJson(fname):
    for line in open(fname, 'r'):
        yield eval(line)
```

```python
data = list(parseJson('Sarcasm_Headlines_Dataset.json'))
```

```python
data
```

```python
len(data)
```

```python
data[0]
```

The dataset consists of 28,619 samples

Each sample in the dataset is a dictionary consisting of 3 key value pairs

'is_sarcastic': label

'headline': text

'article_link': url link to the article

```python
labels=[]
headlines=[]
article_link=[]
```

```python
for item in data:
    labels.append(item['is_sarcastic'])
    headlines.append(item['headline'])
    article_link.append(item['article_link'])
```

```python
labels
```

```python
headlines
```

```python
import tensorflow as tf
```

```python
from tensorflow import keras
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

Tokenization

```python
tokenizer=Tokenizer(oov_token='OOV')
```

```python
tokenizer.fit_on_texts(headlines)
```

```python
word_index=tokenizer.word_index
```

```python
word_index
```

```python
len(word_index)
```

Generate sequences

```python
sequences=tokenizer.texts_to_sequences(headlines)
```

```python
sequences
```

Generated sequences are of different length, in order to feed the sequences into the neaural network, pad the sequences with zeros at the end of the sequences.

Now the length of the sequences is equal to the longest sequence in the list of sequences

```python
padded=pad_sequences(sequences,padding='post')
```

```python
padded.shape
```
